```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('diamonds')
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df['a'] = (df['x'] + df['y'] + df['z'])/3
```

```python
df = df.drop(columns=['x','y','z'])
```

```python
df.hist(bins=50,figsize=(12,8));
```

```python
df.isnull().sum()
```

```python
plt.hist(df['price'])
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("The Distribution of Price")
plt.show()
```

```python
plt.hist(df['carat'])
plt.xlabel("Carat")
plt.ylabel("Frequency")
plt.title("The Distribution of Carat")
plt.show()
```

```python
df['cut'].value_counts()
```

```python
label = ("Ideal","Premium","Very Good","Good","Fair")
plt.pie(df['cut'].value_counts(),explode=[0.03,0,0.1,0.1,0],autopct="",labels=label)
plt.title("The diamond cut")
plt.show()
```
