# Data Cleaning

```python
# Dropping columns with less relevance
df = df.drop(columns='last_review')
df = df.drop(columns='reviews_per_month')
```

```python
# Applying One-Hot Encoding to object dtype values
le = LabelEncoder()
df['name'] = le.fit_transform(df['name'])
df['host_name'] = le.fit_transform(df['host_name'])
df['neighbourhood_group'] = le.fit_transform(df['neighbourhood_group'])
df['neighbourhood'] = le.fit_transform(df['neighbourhood'])
df['room_type'] = le.fit_transform(df['room_type'])
```

# Visualizing the geographical data

```python
df.plot(kind="scatter", x="longitude", y="latitude", grid=True)
plt.show()
```

```python
# Looking for correlation 
corr_matrix = df.corr(numeric_only=True)
corr_matrix["price"].sort_values(ascending=False)
```

```python
# Dropping outliers 
from sklearn.ensemble import IsolationForest

isolation_forest = IsolationForest(random_state=42)
outlier_pred = isolation_forest.fit_predict(df)
```
